```python
import pandas as pd
import numpy as np

movies = pd.read_csv('/content/drive/MyDrive/colab_data_file/movies.csv')
movies
```

```python
rating = pd.read_csv('/content/drive/MyDrive/colab_data_file/ratings.csv')
rating = rating[['userId', 'movieId', 'rating']]
rating
```

```python
rating_matrix = rating.pivot_table('rating', index = 'userId', columns = 'movieId')
rating_matrix
```

```python
rating_movie = pd.merge(rating, movies, on = 'movieId')
rating_movie
```

```python
rating_matrix = rating_movie.pivot_table('rating', index = 'userId', columns = 'title')
rating_matrix
```

```python
rating_matrix = rating_matrix.fillna(0)
rating_matrix = rating_matrix.T
```

```python
rating_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity

item_sim = cosine_similarity(rating_matrix, rating_matrix)
item_sim
```

```python
item_sim_df = pd.DataFrame(data = item_sim, index = rating_matrix.index, columns = rating_matrix.index)
item_sim_df
```

```python
item_sim_df['(500) Days of Summer (2009)'].sort_values(ascending=False) #(500) Days of Summer (2009), [REC] (2007)
```
