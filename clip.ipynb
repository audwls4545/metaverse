```python
from PIL import Image
import requests
```

```python
from transformers import CLIPProcessor, CLIPModel
```

```python
model = CLIPModel.from_pretrained("openai/clip-vit-large-patch14")
```

```python
precessor = CLIPProcessor.from_pretrained("openai/clip-vit-large-patch14")
```

```python
url = 'https://t1.daumcdn.net/news/202301/01/YouTextLife/20230101090105477twfk.jpg'
image = Image.open(requests.get(url, stream=True).raw)

inputs = precessor(text=["a photo of a cat", "a photo of a fish"], images=image, return_tensors="pt", padding=True)

result = model(**inputs)
logits_per_image = result.logits_per_image
```

```python
logits_per_image
```

```python
probs = logits_per_image.softmax(dim=1)
probs
```
