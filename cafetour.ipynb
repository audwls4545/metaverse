```python
f = open('tour.txt','r')
txt = f.read()
f.close()
print(txt)
```

```python
from konlpy.tag import Okt
okt = Okt()

listdata = okt.nouns(txt)
listdata
```

```python
listdata1 = ' '.join(listdata)
listdata1
```

```python
from nltk import Text
from nltk.tokenize import RegexpTokenizer
import matplotlib.pyplot as plt
import numpy as np

plt.rc('font', family='NanumBarunGothic')

retokenize = RegexpTokenizer("[\w]+")
text = Text(retokenize.tokenize(listdata1))
plt.title('상위 30개')
text.plot(30)
plt.show()
```

```python
text.dispersion_plot(['제주','베트남'])
```

```python
#konlpy를 가지고 명사로 추출된 리스트를 dict형태로 변경
word_count = {}
for temp in listdata:
  if temp in word_count:
    word_count[temp] += 1
  else:
    word_count[temp] = 1

word_count
```

```python
from wordcloud import WordCloud
wc = WordCloud(font_path='NanumGothic.ttf', background_color='white',
               width=1000, height=400, max_font_size=500, max_words=1000)

cloud = wc.generate_from_frequencies(word_count)  ##{제주:100, 베트남:50}
cloud.to_image()
```
