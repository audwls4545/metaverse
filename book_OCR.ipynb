```python
from google.colab import files

img_file = files.upload()

print(img_file)
```

```python
img_path = list(img_file.keys())[0]

img_path
```

```python
from PIL import Image

img = Image.open(img_path)
img
```

```python
from paddleocr import PaddleOCR

ocr = PaddleOCR(lang='korean')

result = ocr.ocr(img_path , cls=False)

result
```

```python
print(result[0][0])
```

```python
print(result[0][0][1])
```

```python
print(result[0][0][1][0])
```

```python
len(result[0])
```

```python
speech = ''

for temp in range(len(result[0])):
    speech += result[0][temp][1][0] + ' '

speech
```
