```python
import matplotlib.pyplot as plt
plt.rc('font',family='NanumBarunGothic')
```

```python
import pandas as pd
df = pd.read_csv('서울특별시 공공자전거 대여이력 정보_22.06.csv', encoding='cp949')
df.describe()
```

```python
rental_dict = {}

for temp in df['대여 대여소명']:
    rental = temp.split()[0]

    if rental in rental_dict.keys():
        rental_dict[rental] = rental_dict[rental] + 1
    else:
        rental_dict[rental] = 1

print(rental_dict)
```

```python
rental_df = pd.DataFrame.from_dict([rental_dict])
rental_df
```

```python
rental_df = rental_df.T #컬럼이랑 인덱스 서로 체인지
rental_df
```

```python
rental_df = rental_df.sort_values(by=0)
rental_df.plot(kind='bar')
plt.show()
```
