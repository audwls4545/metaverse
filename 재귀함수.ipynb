# 함수에서 재귀호출 사용

```python
def hello():
    print('Hello, world!')
    hello()
 
hello()
```

재귀호출과 스택 넘침 현상.png

```python
def hello(count):
    if count == 0:    # 종료 조건을 만듦. count가 0이면 다시 hello 함수를 호출하지 않고 끝냄
        return
    
    print('Hello, world!', count)
    
    count -= 1      # count를 1 감소시킨 뒤
    hello(count)    # 다시 hello에 넣음
 
hello(5)    # hello 함수 호출
```

# 재귀 호출로 팩토리얼 구하기

```python
def factorial(n):
    if n == 1:      # n이 1일 때
        return 1    # 1을 반환하고 재귀호출을 끝냄
    return n * factorial(n - 1)    # n과 factorial 함수에 n - 1을 넣어서 반환된 값을 곱함
 
print(factorial(5))
```

```python
num = 0

def factorial(n):
    if n == 1:      # n이 1일 때
        return 1    # 1을 반환하고 재귀호출을 끝냄
    num = n * factorial(n - 1)    
    return num    # n과 factorial 함수에 n - 1을 넣어서 반환된 값을 곱함
 
print(factorial(5))
```

```python
input_list = [1,2,3,4,5]
used = [0]*len(input_list)

def perm(arr, n):
    if n == len(input_list):
        #print(arr)
        return
    for i in range(len(input_list)):
        if not used[i]:
            used[i] = 1
            arr.append(input_list[i])
            print(i)
            print("여기 실행중",arr)
            perm(arr, n+1)
            print("여기 실행중실행중 -> 재귀 종료",arr)
            arr.pop()
            print("여기 실행중실행중실행중 -> 맨 뒤 제거",arr)
            used[i] = 0
            print("used[%d] 0으로 변경"%i)

perm([], 0)
```
