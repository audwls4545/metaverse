```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/colab_data_file/tmdb_5000_movies.csv')
df
```

```python
#문자로 감싸진 것을 다시 리스트로 만들기
from ast import literal_eval

literal_eval(df.iloc[0]['genres'])
```

```python
movies_df = df[['id', 'title', 'genres', 'vote_average',
                'vote_count', 'popularity', 'keywords', 'overview']]
```

```python
movies_df[['genres', 'keywords']]
```

```python
movies_df['genres'] = movies_df['genres'].apply(literal_eval) #문자로 된거를 리스트로 넣어주기
movies_df['keywords'] = movies_df['keywords'].apply(literal_eval)
```

```python
movies_df['genres']
```

```python
name_result = []

for temp in movies_df['genres']:
    names = []
    for temp1 in temp:
        names.append(temp1['name'])
    name_result.append(names)

movies_df['genres'] = name_result
movies_df['genres']
```

```python
key_result = []

for temp in movies_df['keywords']:
    keys = []
    for temp1 in temp:
        keys.append(temp1['name'])
    key_result.append(keys)

movies_df['keywords'] = key_result
movies_df['keywords']
```

```python
data_result = []

for temp in movies_df['genres']:
    data = (' ').join(temp)
    data_result.append(data)

movies_df['genres_literal'] = data_result
movies_df['genres_literal']
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vect = CountVectorizer(min_df = 0, ngram_range = (1, 2))
genre_mat = count_vect.fit_transform(movies_df['genres_literal'])
print(genre_mat.toarray())
print(genre_mat.shape)
count_vect.vocabulary_
```

```python
from sklearn.metrics.pairwise import cosine_similarity
genre_sim = cosine_similarity(genre_mat, genre_mat)
genre_sim
```

```python
genre_sim.argsort()[:,::-1] #내림차순
```

```python
def find_sim_movie(df, sorted_ind, title_name, top_n = 10):
    title_movie = df[df['title'] == title_name]
    title_index = title_movie.index.values
    similar_indexes = sorted_ind[title_index, :(top_n)]

    return df.iloc[similar_indexes[0]]
```

```python
genre_sorted = genre_sim.argsort()[:,::-1]

similar_movies = find_sim_movie(movies_df, genre_sorted, 'The Godfather', 10)
similar_movies['title']
```
