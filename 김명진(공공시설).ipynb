```python
#의료 데이터
#보건소(보건의료원포함) 수
df_medical_infra = pd.read_csv('/content/drive/MyDrive/colab_data_file/보건소_보건지소_보건진료소_수_20230605233927.csv',encoding='euc-kr')
df_medical_infra.head()
```

```python
data_medical_infra = df_medical_infra.copy() #복제

data_medical_infra = data_medical_infra.rename(columns={'시도별(1)':'sido', '2021.1':'medical'}) #컬럼명 변경
data_medical = data_medical_infra.loc[:,['sido','medical']] #원하는 부분만 선택
data_medical.head()
```

```python
data_medical= data_medical.drop(data_medical.index[0:2]) #필요없는 2개 행 제거
data_medical.head()
```

```python
data_medical = data_medical.astype({'medical':int}) #정수데이터와 문자데이터 혼합되어있어서 정수로 통일
data_medical = data_medical.sort_values(by='medical', ascending=False) # 내림차순
data_medical = data_medical.reset_index() #인덱스 새로 만들기
data_medical = data_medical.drop('index', axis=1) # index열 제거
data_medical = data_medical.head(10) #상위 10개만 출력
data_medical
```
